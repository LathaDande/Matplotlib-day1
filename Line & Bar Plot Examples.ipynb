```python
from matplotlib import pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
z=[4,3,2,1]
a=[3,2,1,4]
plt.plot(x,y,ls="dotted",lw=2,c="red",marker="o",mec="blue",mfc="yellow",ms=10)
plt.plot(x,z,ls="dotted",lw=2,c="blue",marker="X",mec="green",mfc="brown",ms=10)
plt.plot(x,a,ls="dotted",lw=2,c="yellow",marker="D",mec="red",mfc="blue",ms=10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
lab=["Line Plot-1","Line Plot-2","Line Plot-3"]
plt.legend(labels=lab,loc="best")
plt.grid()
plt.xlim([0,5])
plt.ylim([0,5])
plt.title("Line Graph")
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
z=[4,3,2,1]
a=[3,2,1,4]
plt.subplot(2,2,1)
plt.plot(x,y,ls="dotted",lw=2,c="red",marker="o",mec="blue",mfc="yellow",ms=10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph-1")
plt.subplot(2,2,2)
plt.plot(x,z,ls="dotted",lw=2,c="blue",marker="X",mec="green",mfc="brown",ms=10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph-2")
plt.subplot(2,2,3)
plt.plot(x,a,ls="dotted",lw=2,c="yellow",marker="D",mec="red",mfc="blue",ms=10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph-3")
plt.subplot(2,2,4)
plt.plot(x,a,ls="dotted",lw=2,c="yellow",marker="D",mec="red",mfc="blue",ms=10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph-4")
plt.suptitle("All plots in one Axis")
```
